# Import libraries

```python
import pandas as pd
import numpy as np
import random
```

# Import csv

# Create lost_dataframe solutions

```python
# generate a random list of unique numbers
def generateRandomList(max, nbr):
    return random.sample(range(1, max+1), nbr)
```

```python
# create a dataframe with the same structure as df
lostDf = pd.DataFrame(columns=df.columns)
# for each row in the dataframe, we create 4 new row in the lostDf
for index, row in df.iterrows():
    for i in range(1):
        newRow = {'Date': object, 'N1': int, 'N2': int, 'N3': int, 'N4': int, 'N5': int, 'E1': int, 'E2': int}
        newRow['Date'] = row['Date']
        # generate a list of random numbers
        randomList = generateRandomList(50, 5)
        # check if randomList is different from the current row
        while (randomList == row[1:6].tolist()):
            randomList = generateRandomList(50, 5)
        newRow['N1'] = randomList[0]
        newRow['N2'] = randomList[1]
        newRow['N3'] = randomList[2]
        newRow['N4'] = randomList[3]
        newRow['N5'] = randomList[4]
        # generate a list of random "etoile" numbers
        randomList = generateRandomList(12, 2)
        # check if randomList is different from the current row
        while (randomList == row[6:8].tolist()):
            randomList = generateRandomList(12, 2)
        newRow['E1'] = randomList[0]
        newRow['E2'] = randomList[1]
        lostDf = lostDf.append(newRow, ignore_index=True)
```

```python
# define the gain and winner column
lostDf['Gain'] = 0
lostDf['Winner'] = 0
```

```python
lostDf
```

# Merge 2 dataframes

```python
# concatenate df and lostDf
finalDf = pd.concat([df, lostDf], axis=0)
# sort the dataframe by date
finalDf = finalDf.sort_values(by=['Date', 'Gain']).reset_index(drop=True)
```
